## 🧩 Paso 3: Limpieza básica de datos

### 3.1 Corregir sentinels y fechas imposibles

**🎯 Objetivo:**
Aplicar reglas de limpieza para reemplazar valores sentinels y corregir fechas imposibles.

**Instrucciones:**
- En `age`, reemplaza el sentinel **-999** con la mediana.
- En `city`, reemplaza el sentinel `?` por valores nulos (`pd.NA`).
- Marca como nulas (`pd.NA`) las fechas fuera de rango.


```python
users_cleaned = users.copy()
usage_cleaned = usage.copy()

users_cleaned['age'] = users_cleaned['age'].replace(-999, np.nan)
users_cleaned.loc[users_cleaned['age'] == 0, 'age'] = np.nan
age_mediana = users_cleaned['age'].median()
users_cleaned['age'] = users_cleaned['age'].fillna(age_mediana)

users_cleaned['city'] = users_cleaned['city'].replace('?', pd.NA)
users_cleaned['plan'] = users_cleaned['plan'].astype(str).str.lower().str.strip()
for old, new in [('á','a'), ('é','e'), ('í','i'), ('ó','o'), ('ú','u')]:
    users_cleaned['plan'] = users_cleaned['plan'].str.replace(old, new, regex=False)

users_cleaned['reg_date'] = pd.to_datetime(users_cleaned['reg_date'], errors='coerce')
users_cleaned.loc[users_cleaned['reg_date'] > '2024-12-31', 'reg_date'] = pd.NaT

print('Limpieza completada para users:')
print(f"age nulos: {users_cleaned['age'].isna().sum()}")
print(f"city nulos: {users_cleaned['city'].isna().sum()}")
print(f"reg_date nulos: {users_cleaned['reg_date'].isna().sum()}")
```

## 🧩 Paso 4: Summary statistics de uso por usuario

### 4.1 Agrupación por comportamiento de uso

🎯 Objetivo: Resumir las variables clave de la tabla `usage` por usuario, creando métricas que representen su comportamiento real de uso histórico.

**Instrucciones:**
1. Construye una tabla agregada de `usage` por `user_id` que incluya:
- número total de mensajes
- número total de llamadas
- total de minutos de llamadas

2. Renombra las columnas para que tengan nombres claros:
- `cant_mensajes`
- `cant_llamadas`
- `total_minutos_llamada`
3. Combina esta tabla con `users`.


```python
usage_cleaned['is_text'] = (usage_cleaned['type'] == 'text').astype(int)
usage_cleaned['is_call'] = (usage_cleaned['type'] == 'call').astype(int)

usage_agg = usage_cleaned.groupby('user_id').agg(
    cant_mensajes=('is_text', 'sum'),
    cant_llamadas=('is_call', 'sum'),
    total_minutos_llamada=('duration', 'sum')
).reset_index()

user_profile = users_cleaned.merge(usage_agg, on='user_id', how='left')
user_profile[['cant_mensajes', 'cant_llamadas', 'total_minutos_llamada']] = user_profile[['cant_mensajes', 'cant_llamadas', 'total_minutos_llamada']].fillna(0)

print('Resumen de uso por usuario:')
display(usage_agg.head())
print('Perfil de usuario combinado:')
display(user_profile.head())
```

- Usa un bucle para generar los 4 boxplots de forma automática.
- Calcula los límites con el método IQR cuando sea necesario.
- Decide si los outliers deben mantenerse, corregirse o revisarse más a fondo.


<div class="alert alert-block alert-success">
<b>Correcciones aplicadas</b>

## 🧩 Paso 6: Segmentación de Clientes

### 6.1 Segmentación de Clientes Por Uso

🎯 Objetivo: Clasificar a cada usuario en un grupo de uso basado en llamadas y mensajes.

**Instrucciones:**
- Crea la columna `grupo_uso` en el dataframe `user_profile`.
- Asigna:
  - `Bajo uso` cuando llamadas < 5 y mensajes < 5
  - `Uso medio` cuando llamadas < 10 y mensajes < 10
  - `Alto uso` para el resto de casos


```python
usage_limits = [
    (user_profile['cant_llamadas'] < 5) & (user_profile['cant_mensajes'] < 5),
    (user_profile['cant_llamadas'] < 10) & (user_profile['cant_mensajes'] < 10)
]
choices = ['Bajo uso', 'Uso medio']
user_profile['grupo_uso'] = np.select(usage_limits, choices, default='Alto uso')

print(user_profile['grupo_uso'].value_counts())
```
